```python
import pandas as pd
```

```python
df_json=pd.read_csv('/content/json_clear.csv')
df_csv=pd.read_csv('/content/csv_clear.csv', dtype = {'TicketNumber': str})
df_yaml=pd.read_csv('/content/yaml_clear.csv')
df_tab=pd.read_csv('/content/tab_cleared.csv', dtype = {'ProgrammNumber': str, 'TicketNumber': str})
df_xml=pd.read_csv('/content/xml_cleared.csv')
df_xlsx=pd.read_csv('/content/xlsx_clear.csv')
```

#JSON + YAML

```python
df_yaml=df_yaml.drop('Unnamed: 0', axis = 1)
df_yaml['ProgrammNumber']=df_yaml['ProgrammNumber'].astype('str')
df_yaml['DepartureDate']=df_yaml['DepartureDate'].astype('datetime64[ns]')
```

```python
df_json=df_json.drop('Unnamed: 0', axis = 1)
df_json['ProgrammNumber']=df_json['ProgrammNumber'].astype('int')
df_json['ProgrammNumber']=df_json['ProgrammNumber'].astype('str')
df_json['DepartureDate']=df_json['DepartureDate'].astype('datetime64[ns]')
```

```python
df_json
```

```python
df_yaml
```

```python
unit_yaml_json = pd.merge(df_yaml, df_json, on=['ProgrammNumber','Programm', 'FlightNumber', 'AirportArrival', 'AirportDeparture','DepartureDate'], how='outer', suffixes=('_yaml', '_json'))
```

```python
unit_yaml_json
```

```python
unit_yaml_json.to_csv('merge_yaml_json.csv')
```

```python
coords_dep = pd.read_csv('GlobalAirportDatabase.txt', sep=':', header=None)[[1,14,15]].rename(columns={1:'AirportDeparture', 14: 'LatitudeDeparture', 15:'LongitudeDeparture'}).dropna().drop_duplicates('AirportDeparture')
coords_arr = pd.read_csv('GlobalAirportDatabase.txt', sep=':', header=None)[[1,14,15]].rename(columns={1:'AirportArrival', 14: 'LatitudeArrival', 15:'LongitudeArrival'}).dropna().drop_duplicates('AirportArrival')
```

```python
unit_yaml_json = pd.merge(unit_yaml_json, coords_arr, on = ['AirportArrival'], how='left')
unit_yaml_json = pd.merge(unit_yaml_json, coords_dep, on = ['AirportDeparture'], how='left')
```

```python
unit_yaml_json.to_csv('merge_yaml_json.csv')
```

```python
unit_tab_json_yaml = pd.merge(unit_yaml_json, df_tab, on = ['DepartureDate', 'FlightNumber', 'ProgrammNumber', 'Programm', 'AirportDeparture','Class', 'AirportArrival', 'FirstName', 'LastName' ],how='outer', suffixes=('_yaml_json', '_tab') )
```

```python
unit_tab_json_yaml.to_csv('merge_tab_json_yaml.csv')
```
